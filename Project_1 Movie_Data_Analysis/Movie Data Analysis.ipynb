# Movie Data Analysis Project

**Project Overview:**

In this data analysis project, I delve into a comprehensive dataset from the world of cinema. The dataset contains information about thousands of movies, including details on their genres, budgets, revenues, release years, popularity, and more. I aim to explore and draw valuable insights from this dataset, shedding light on various aspects of the movie industry.

**Key Questions:**

To guide my analysis, I have defined specific questions and objectives, such as:

1. What are the most common movie genres in the dataset?
2. Which directors have produced the highest number of movies?
3. How has movie production evolved over the years?
4. What is the distribution of vote averages among the movies?
5. Is there a correlation between movie budgets and revenues?




```python
import pandas as pd  # For data handling and manipulation
import numpy as np   # For numerical operations
import matplotlib.pyplot as plt  # For data visualization
import seaborn as sns  # For enhanced data visualization
import datetime  # For date-related operations
```

```python
df= pd.read_csv("tmdb-movies.csv")
df.head(20)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df.info()
```

```python

# Remove duplicate records
df.drop_duplicates(inplace=True)

# Handle missing values
df['imdb_id'].fillna('', inplace=True)
df['cast'].fillna('', inplace=True)
df['homepage'].fillna('', inplace=True)
df['director'].fillna('', inplace=True)
df['tagline'].fillna('', inplace=True)
df['keywords'].fillna('', inplace=True)
df['overview'].fillna('', inplace=True)
df['genres'].fillna('', inplace=True)
df['production_companies'].fillna('', inplace=True)

# Convert 'release_date' to datetime
df['release_date'] = pd.to_datetime(df['release_date'], format='%m/%d/%y')

# Verify data types and non-null counts
data_info = df.info()

```

```python
# Describe all columns in the dataset
df.describe(include='all')
```

```python
# Explore movie genres
genres = df['genres'].str.cat(sep='|').split('|')
unique_genres = set(genres)
print("Unique Genres:", unique_genres)
```

```python
# Explore the most common genres
genre_counts = df['genres'].str.split('|', expand=True).stack().value_counts()
print("Most Common Genres:")
print(genre_counts.head(10))
```

```python
# Analyze the distribution of vote averages
plt.figure(figsize=(8, 5))
sns.histplot(df['vote_average'], kde=True)
plt.title('Distribution of Vote Averages')
plt.xlabel('Vote Average')
plt.ylabel('Count')
plt.show()
```

```python
votes = df['vote_average'].value_counts()
votes
```

```python
# Analyze the distribution of movie release years
plt.figure(figsize=(12, 6))
sns.histplot(df['release_year'], bins=30, kde=False)
plt.title('Distribution of Movie Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

```python
release_year = df['release_year'].value_counts()
release_year
```

```python
# Explore the relationship between budget and revenue
plt.figure(figsize=(8, 8))
sns.scatterplot(x='budget', y='revenue', data=df)
plt.title('Budget vs. Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

```python
correlation_budget_revenue = df['budget'].corr(df['revenue'])
print('correlation_budget_revenue')
correlation_budget_revenue
```

```python
# Explore the top directors with the most movies
top_directors = df['director'].value_counts().head(10)
print("Top Directors with the Most Movies:")
print(top_directors)
```

```python
# Explore the relationship between popularity and revenue
plt.figure(figsize=(8, 8))
sns.scatterplot(x='popularity', y='revenue', data=df)
plt.title('Popularity vs. Revenue')
plt.xlabel('Popularity')
plt.ylabel('Revenue')
plt.show()
```

```python
correlation_popularity_revenue = df['popularity'].corr(df['revenue'])
print('correlation_popularity_revenue')
correlation_popularity_revenue
```

## Conclusion

Based on the analysis of the IMDB dataset, several insights and findings have been revealed:

**Genres:**
- The dataset contains a variety of movie genres, with Drama being the most common, followed by Comedy, Thriller, Action, and Romance.
- Drama and Comedy are the two most prevalent genres in the dataset, with Drama being significantly more common than other genres.

**Directors:**
- Woody Allen is the director with the most movies in the dataset, with 45 movies attributed to him.
- Several directors, such as Clint Eastwood, Martin Scorsese, and Steven Spielberg, have a substantial number of movies to their names as well.

**Release Years:**
- The dataset spans a wide range of release years, with the most movies released in 2014, 2013, and 2015.
- The number of movies released in recent years has been consistently high.

**Vote Average:**
- The vote average distribution is relatively symmetric, with a peak around 6.0.
- The dataset contains movies with a wide range of vote averages, from as low as 2.0 to as high as 9.2.

**Correlation:**
- The correlation between a movie's budget and its revenue is positive and relatively strong, with a coefficient of approximately 0.735. This suggests that there is a significant relationship between a movie's budget and its revenue, indicating that higher-budget movies tend to earn higher revenues.
- The correlation between a movie's popularity and its revenue is also positive, with a coefficient of around 0.663. This implies that movies with higher popularity tend to generate higher revenue.

These findings provide valuable insights for filmmakers, studios, and movie enthusiasts. Understanding the relationship between factors such as genre, director, release year, and vote average can help in making informed decisions regarding movie production and distribution. Additionally, the strong correlation between budget and revenue emphasizes the financial aspect of the movie industry, highlighting the importance of budget allocation for potential revenue generation.

These insights can inform decisions related to movie production, marketing, and distribution, and can serve as a foundation for further analysis and decision-making in the film industry.
